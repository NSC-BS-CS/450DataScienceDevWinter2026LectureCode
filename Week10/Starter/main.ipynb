## Renaming Messy Columns

Convention says:

- Column names should be lowercase.
- Whitespace should be replaced with underscores (snake_case).



```python
df_columns_renamed= df.rename(columns = {
  "Serial": "serial_id",
    "FireCategory": "fire_category",
    "FireYear": "fire_year",
    "Area": "area_name",
    "DistrictName": "district_name",
    "UnitName": "unit_name",
    "FullFireNumber": "full_fire_number",
    "FireName": "fire_name",
    "EstTotalAcres": "estimated_total_acres",
    "ReportDateTime": "report_datetime",
    "CreationDate": "created_at",
    "ModifiedDate": "modified_at",
    "DistrictCode": "district_code",
    "UnitCode": "unit_code",
    "DistFireNumber": "district_fire_number"
})

df.columns = df_columns_renamed.columns.str.lstrip().str.rstrip().str.replace(" ", "_").str.lower()

```

## Removing NaN

For the most part, removing NaN values is a good thing because it removes missing data that impacts your analysis.

You must decide whether you are:
- Removing an entire row if it contains a NaN
- Removing columns that contain a large amount of NaN values
- Remove only rows or columns that are completely NaN
Consider
- The amount of data lost
- Does the missing data have meaning
- Should you fill the NaN instead of removing the row completely 

```python
print(df.shape)
df = df.dropna(how="all", axis="columns")
df = df.dropna(how="all")
df = df.drop(columns=["cause_comments"])
print(df.isna().sum().sort_values(ascending=False))
```

```python
df["specificcause"].unique()
```

```python
df["causeby"].unique()
```

```python
df["specificcause"] = df["specificcause"].fillna("unknown")
df["causeby"] = df["causeby"].fillna("unknown")
print(df.isna().sum().sort_values(ascending=False))
```

```python
df = df.dropna()
df.shape
```

## Updating Dates

You must convert your dates from strings into datetime objects.  
This allows you to use built-in date methods and perform time-based calculations.


```python
df["ign_datetime"].head(20)
df["discover_datetime"].head(20)
df["control_datetime"].head(20)
df["report_datetime"].head(20)
```

```python
date_columns = [
    "ign_datetime",
    "discover_datetime",
    "control_datetime",
    "report_datetime"
]

for column in date_columns:
    df[column] = pd.to_datetime(
        df[column],
        format="%Y/%m/%d %H:%M:%S %p",
        errors="coerce"
    )
df[[
    "ign_datetime",
    "discover_datetime",
    "control_datetime",
    "report_datetime"
]].dtypes
```

```python
df["discovery_to_control"] = df["control_datetime"] - df["discover_datetime"]
df["discovery_to_control"]
```

## GroupBy
- `groupby()`
Another method inspired by the SQL command of the same name, `groupby()` allows us to group data based on specific parameters and then perform calculations on those groups.

Think of it like sorting data into buckets based on a column value. Let's say we have data on t-shirts and we group by size. Each row would be placed into a bucket (group) based on the size value.

For example, if we grouped by `size`, all rows with `size = S` would go into one bucket, all rows with `size = M` into another, and all rows with `size = L` into another.

Once the data is organized into these groups, we can perform calculations on each bucket. For instance, we could count how many t-shirts are in each size, calculate the average price per size, or find the total number of shirts sold for each size.

## Aggregate Methods

The methods we've been using are **aggregation methods**. We've used them before, but not yet in combination with `groupby()`.

To review, some common aggregation methods include:

- `sum()` – total values  
- `mean()` – average values  
- `count()` – number of rows  
- `nunique()` – number of unique values  
- `min()` – smallest value  
- `max()` – largest value
